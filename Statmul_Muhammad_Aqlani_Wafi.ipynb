```python
#Transformation Categorical Data to Numerical Data
df4 = df3.iloc[:,0:]
# df4['Q23'] = df['Q23'].replace('Currently not employed', 0)
df4.iloc[:,2:91] = df4.iloc[:,2:91].notna()
df4.iloc[:,2:91] = df4.iloc[:,2:91].astype('int')
print("\nNilai unik sebelum perubahan:")
print(df4['Q23'].astype('category').cat.categories)
data = df4
# Converting type of columns to category
# df4['Q11'] = df4['Q11'].astype('category').cat.codes
# df4['Q16'] = df4['Q16'].astype('category').cat.codes
# df4['Q23'] = df4['Q23'].astype('category').cat.codes
# enc = OneHotEncoder()

# # Passing encoded columns
# enc_data = pd.DataFrame(enc.fit_transform(
# 	df4[['Q11','Q23']]).toarray())

# Merge with main
# New_df = data.join(enc_data)
New_df = data
print("\nNilai unik setelah perubahan:")
print(df4['Q23'].unique())
New_df
```

```python
DA = New_df[New_df['Q23']=='Data Architect']
DA['Q11'] = DA['Q11'].astype('category').cat.codes
DA['Q16'] = DA['Q16'].astype('category').cat.codes
DA['Q23'] = DA['Q23'].astype('category').cat.codes
DA
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Misalkan df_encoded adalah DataFrame yang sudah di-encode (One-Hot Encoding)
# Pastikan data sudah dalam bentuk numerik sebelum PCA
DA1 = DA.iloc[:,:91]
# Menstandarisasi data menggunakan StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(DA1)

# Terapkan PCA pada data
pca = PCA()
pca.fit(df_scaled)

# Menampilkan varians yang dijelaskan oleh setiap komponen
explained_variance = pca.explained_variance_ratio_

# Plotting scree plot
plt.figure(figsize=(8,6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.axhline(y=0.05, color='red', linestyle='-')
plt.title('Scree Plot')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid(True)
plt.show()
```

```python
# Misalkan kamu sudah melakukan PCA dan memilih 2 komponen utama
pca = PCA(n_components=7)
df_pca = pca.fit_transform(DA1)

# Membuat DataFrame untuk komponen utama
df_pca_df = pd.DataFrame(data=df_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'])
DA.reset_index(drop=True, inplace=True)
df_pca_df.reset_index(drop=True, inplace=True)
# Menggabungkan komponen utama ke dataset asli jika diperlukan
df_final = pd.concat([DA, df_pca_df], axis=1)

# Sekarang kamu bisa melakukan MANOVA pada data yang sudah digabungkan
from statsmodels.multivariate.manova import MANOVA
formula = 'Q16 + Q11 ~ PC1 + PC2 + PC3 + PC4 + PC5 + PC6 + PC7'
manova = MANOVA.from_formula(formula, data=df_final)
# print(manova.fit())
print(manova.mv_test())
```

**PC1 (Komponen Utama 1):**

Semua tes (Wilks' Lambda, Pillai's Trace, Hotelling-Lawley,Roy's Greatest Root) menunjukkan nilai 0.0000 untuk p-value. Ini menunjukkan bahwa ada perbedaan yang sangat signifikan antara grup yang dianalisis untuk komponen utama pertama. F-value yang sangat tinggi (11893.0927) menunjukkan adanya variasi yang besar antara grup pada PC1.

Kesimpulan: Komponen pertama ini sangat berpengaruh terhadap perbedaan antara grup dalam data.

**PC2 (Komponen Utama 2):**

Hasilnya juga sangat signifikan dengan p-value yang sangat kecil (0.0000).
F-value (152.2199) lebih kecil daripada PC1, tetapi masih menunjukkan bahwa komponen ini penting untuk membedakan grup.

Kesimpulan: Komponen kedua juga memberikan kontribusi signifikan terhadap perbedaan antara grup.

**PC3 hingga PC7:**

Semua komponen dari PC3 hingga PC7 menunjukkan p-value yang sangat kecil (0.0000), yang berarti masing-masing komponen ini juga berkontribusi signifikan terhadap perbedaan grup.

F-value untuk setiap komponen juga menunjukkan bahwa perbedaan antar grup tetap signifikan, meskipun sedikit lebih rendah dibandingkan dengan PC1 dan PC2.

**Kesimpulan:**

Semua komponen utama (PC1 hingga PC7) memberikan kontribusi signifikan terhadap perbedaan antar grup, dengan nilai p-value yang sangat kecil.

PC1 dan PC2 memberikan kontribusi yang lebih besar terhadap variasi antar grup, sementara PC3 hingga PC7 juga penting meskipun kontribusinya sedikit lebih kecil.
